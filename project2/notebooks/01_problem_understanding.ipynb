# Hiểu Vấn Đề

- **Mục tiêu kinh doanh**: Xây dựng hệ thống gợi ý xe máy tương tự và phân khúc sản phẩm cho nền tảng thương mại điện tử (ví dụ Chợ Tốt).
- **Các câu hỏi chính**:
  - Làm thế nào đề xuất được các xe máy tương tự cho một mẫu xe cụ thể?
  - Các phân khúc xe máy khác nhau biểu hiện ra sao về giá, hãng, dòng xe, năm sản xuất?
- **Ý nghĩa**: Hỗ trợ người mua tìm xe phù hợp nhanh hơn, và hỗ trợ đội ngũ kinh doanh tối ưu chiến lược định giá.
- **Nguồn dữ liệu**: Tập tin `data_motobikes.xlsx - Sheet1.csv` gồm thông tin mô tả chi tiết cho hơn 7k tin đăng.
