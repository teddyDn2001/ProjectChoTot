# EDA Tự Động

Notebook này tạo báo cáo EDA tự động bằng `ydata_profiling` và `dataprep`.


```python
import pandas as pd
from pathlib import Path
from ydata_profiling import ProfileReport

try:
    from dataprep.eda import create_report
    HAS_DATAPREP = True
except ImportError:
    HAS_DATAPREP = False
    print("[WARN] dataprep chưa được cài đặt. Bỏ qua phần tạo báo cáo dataprep.")

DATA_PATH = Path("/Users/doananh/Documents/Documents - Doan’s MacBook Pro/đồ án DS/project2/data_motobikes.xlsx - Sheet1.csv")
REPORT_DIR = Path("../reports")
REPORT_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(DATA_PATH)

profile = ProfileReport(df, title="Motorbike Listings Profiling Report", minimal=True)
profile.to_file(REPORT_DIR / "profiling_report.html")

if HAS_DATAPREP:
    report = create_report(df)
    report.save(str(REPORT_DIR / "dataprep_report.html"))
else:
    print("Gợi ý: chạy `pip install dataprep` rồi chạy lại cell này để sinh báo cáo dataprep.")
```
